```python
# Second Matplotlib Tutorial
# Uses Numpy,csv, counter

plt.style.use("ggplot")

#renames data2.csv as csv_file and then reads it into variable named csv_reader
with open('datatut2.csv') as csv_file:
    csv_reader = csv.DictReader(csv_file)
    lang_counter = Counter()
    
    #iterates through rows updating language counter class. 
    for row in csv_reader:
        lang_counter.update(row['LanguagesWorkedWith'].split(';'))

languages = []
popularity = []

#most common method creates a list of tuples with (<lang name>, <num of instances>)      
for item in lang_counter.most_common(15):
    #code below then breaks up the tuple into respective lists
    languages.append(item[0])
    popularity.append(item[1])

languages.reverse()
popularity.reverse()
#creating horizontal bar chart with the extrapolated data
plt.barh(languages, popularity)
plt.title('Most Popular Languages')
plt.ylabel('Language Name')
plt.xlabel('Num of Users')
plt.show()

    
```

```python
#Third Matplotlib Tutorial

plt.style.use('seaborn')

#reading data into var
data = pd.read_csv('datatut2.csv')

#breaking data into respective lists
ids = data['Responder_id']
lang_responses = data['LanguagesWorkedWith']

lang_counter = Counter()    
#iterates through rows updating language counter class. 
for response in lang_responses:
    lang_counter.update(response.split(';'))

languages = []
popularity = []
for item in lang_counter.most_common(5):
    #code below then breaks up the tuple into respective lists
    languages.append(item[0])
    popularity.append(item[1])


#Pie Chart creation
explode = [0, 0, 0, 0.1, 0]
plt.pie(popularity, labels = languages, explode = explode, autopct='%1.1f%%')
plt.title('Top 5 Most Popular Langs')

plt.show()
```

```python
# Fifth MatPlotLib Tutorial

#uses Pandas and pyplot

#this library loads in the csv data from github
from urllib.request import urlretrieve
urlretrieve('https://raw.githubusercontent.com/CoreyMSchafer/code_snippets/master/Python/Matplotlib/05-Fill_Betweens/data.csv', 'datatut5.csv')

data = pd.read_csv('datatut5.csv')
ages = data['Age']
dev_salaries = data['All_Devs']
py_salaries = data['Python']
js_salaries = data['JavaScript']

plt.plot(ages, dev_salaries, color='#444444', linestyle='--', label='All Devs')

plt.plot(ages, py_salaries, label='Python')

overall_median = 57287


plt.fill_between(ages, py_salaries,dev_salaries,
                 where=(py_salaries > dev_salaries), 
                 interpolate=True, alpha = 0.25, label='Above Avg')

plt.fill_between(ages, py_salaries, dev_salaries,
                 where=(py_salaries <= dev_salaries), 
                 interpolate=True, color='red', alpha = 0.25, label='Below Avg')

plt.legend()

plt.title('Median Salary (USD) by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')

plt.tight_layout()

plt.show()
```
